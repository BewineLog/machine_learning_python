```python
"""
DataFrame과 리스트 , 딕셔너리, 넘파이 ndarray 상호 변환
"""
#1차원 형태의 데이터 기반 변환

import numpy as np
col_name1 = ['col1']
list1 = [1,2,3]
array1 = np.array(list1)

#list to DataFrame
df_list1 = pd.DataFrame(list1, columns = col_name1)
print("1차원 list로 만든 DataFrame:\n",df_list1,'\n')

#numpy ndarray to DataFrame
df_array1 = pd.DataFrame(array1, columns = col_name1)
print("1차원 ndarray로 만든 DataFrame:\n",df_array1,'\n')
```

```python
#2차원 형태의 데이터 기반 변환 , 2행 3열

col_name2 = ['col1','col2','col3']

list2 = [[1,2,3],
        [11,12,13]]

array2 = np.array(list2)
print('array2 shape:',array2.shape,'\n')

#list to DataFrame
df_list2 = pd.DataFrame(list2, columns = col_name2)
print("2차원 리스트로 만든 DataFrame:\n",df_list2,'\n')

#ndarray to DataFrame
df_array2 = pd.DataFrame(array2, columns = col_name2)
print('2차원 ndarray로 만든 DataFrame:\n',df_array2,'\n')


```

```python
#Dictionary to DataFrame
dict = {'col1': [1,11], 'col2':[2,22], 'col3':[3,33]}

df_dict = pd.DataFrame(dict) # key 값이 column으로 , value는 column data로 적용

print('Dictionary로 만든 DataFrame:\n',df_dict,'\n')
```

```python
#DataFrame to ndarray
array3 = df_dict.values
print('df_dict.values 타입:',type(array3),'df_dict.values shape:',array3.shape)
print(array3)
```

```python
#DataFrame to list
list3 = df_dict.values.tolist()
print('df_dict.values.tolist() 타입:',type(list3))
print(list3,'\n')

#DataFrame to dictionary
dict3 = df_dict.to_dict('list')
print('df_dict.to_dict() 타입:',type(dict3))
print(dict3)
```
