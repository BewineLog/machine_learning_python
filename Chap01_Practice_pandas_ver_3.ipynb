```python
"""
데이터 셀렉션 및 필터링 by pandas
"""

import pandas as pd

titanic_df = pd.read_csv('/Users/ijeonghan/dev/jupyter/train.csv')

#DataFrame 뒤의 [] 연산자
print('단일 칼럼 데이터 추출:\n',titanic_df['Pclass'].head(3))
print('\n여러 칼럼의 데이터 추출:\n',titanic_df[['Survived', 'Pclass']].head(3),'\n')
print('[] 안에 숫자 인덱스는 KeyError 오류 발생:\n',titanic_df[0])
```

```python
#DataFrame의 [] 안에 Boolean indexing
titanic_df[titanic_df['Pclass'] == 3].head()
```

```python
#ix[], iloc[], loc[] 연산자를 위한 DataFrame 생성
data = {'Name': ['Chulmin', 'Eunkyung', 'Jinwoong', 'Soobeom'],
        'Year': [2011, 2016, 2015, 2015],
        'Gender': ['Male', 'Female', ' Male', 'Male']
       }

data_df = pd.DataFrame(data, index =['one', 'two', 'three', 'four'])
data_df
```

```python
#ix[행 위치 지정, 열 위치 지정], 칼럼 명칭 기반 or 칼럼 위치 기반 인덱싱 방식 사용 , 혼란 및 가독성 때문에 사용하지 않는 방식(pandas 1.0.0부터 삭제)

#print('칼럼 위치 기반 인덱싱 데이터 추출:',titanic_df.ix[0,2])
#print('칼럼 명칭 기반 인덱싱 데이터 추출:',titanic_df.ix[0,'Pclass'])
```

```python
#새로운 row index 값 생성

data_df_reset = data_df.reset_index()
data_df_reset = data_df_reset.rename(columns ={'index': 'old_index'})

data_df_reset.index = data_df_reset.index + 1
data_df_reset
```

```python
#iloc[] 연산자, 위치 기반 인덱싱을 사용하는 연산자, 단 Boolean indexing은 제공 x

data_df.iloc[0,0]
```

```python
#Error 유발 코드,  column 값에 명칭 기반 인덱싱
data_df.iloc[0,'Name']
```

```python
#Error 유발 코드, row 값에  명칭기반 인덱싱
data_df.iloc['one',0]
```

```python
# index가 1부터 시작하는 새로운 정수형 인덱스 값을 가지고 있는 data_df_reset에 iloc적용
data_df_reset.iloc[0,1]
```

```python
#DataFrame loc[] 연산자, 명칭 기반 인덱싱 사용하는 연산자 , [DataFrame's index, columns name]
data_df.loc['one', 'Name']
```

```python
#data_df_reset에 설정한 정수형 인덱스 값 활용(index: 1 ~)
data_df_reset.loc[1,'Name']
```

```python
#다음 코드는 error 발생합니다.
data_df_reset.loc[0,'Name']
```

```python
#명칭 기반 인덱싱, 위치 기반 인덱싱에서 슬라이싱 적용 시 차이점 확인
print('위치 기반 iloc slicing\n', data_df.iloc[0:1,0],'\n')
print('명칭 기반 loc slicing\n', data_df.loc['one':'two', 'Name']) # 명칭 기반 인덱싱에서 슬라이싱은 종료점을 포함한 위치를 반환
```

```python
#Boolean indexing
titanic_boolean = titanic_df[titanic_df['Age'] > 60]
print(type(titanic_boolean))
titanic_boolean.head()
```

```python
#Boolean indexing에서 원하는 칼럼 별도 추출하기
titanic_df[titanic_df['Age'] > 60][['Name', 'Age']].head()
```

```python
#Boolean indexing에서 loc 연산자를 통해 칼럼 별도 추출하기
titanic_df.loc[titanic_df['Age'] > 60, ['Name', 'Age']].head()
```

```python
#여러 조건들을 결합하여 데이터 추출하기 - 나이 60세 이상, 선실 1등급, 성별 여성
titanic_df[(titanic_df['Age'] > 60) & (titanic_df['Pclass'] == 1)&
          (titanic_df['Sex'] == 'female')]
```

```python
#여러 조건들을 결합하여 데이터 추출하기2
cond1 = titanic_df['Age'] > 60
cond2 = titanic_df['Pclass'] == 1
cond3 = titanic_df['Sex'] == 'female'

titanic_df[cond1 & cond2 & cond3]
```
