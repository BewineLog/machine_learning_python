```python
"""
-DataFrame의 정렬, Aggregation 함수, GroupBy 적용 & 데이터 가공 및 결손 데이터 처리
"""
import pandas as pd

titanic_df = pd.read_csv('/Users/ijeonghan/dev/jupyter/train.csv')
```

```python
#정렬, sort_values() method 활용
titanic_sorted = titanic_df.sort_values(by = ['Name'])
titanic_sorted.head()
```

```python
#정렬 기준에 여러 개의 칼럼 설정(list 형식) & 내림차순 정렬로 설정
titanic_sorted = titanic_df.sort_values(by = ['Pclass', 'Name'], ascending = False)
titanic_sorted.head()
```

```python
#Aggregation 함수 적용 (min, max, sum, count ... etc) -> 모든 칼럼에 적용된 결과 반환
titanic_df.count()
```

```python
#특정 칼럼에 aggregation 함수 적용
titanic_df[['Age', 'Fare']].mean()
```

```python
#groupby() 적용
titanic_groupby = titanic_df.groupby(by = 'Pclass')
print(type(titanic_groupby))
```

```python
titanic_groupby = titanic_df.groupby('Pclass').count()
titanic_groupby
```

```python
#DataFrameGroupBy 객체에 필터링 적용 & aggregation 함수 적용
titanic_groupby = titanic_df.groupby('Pclass')[['PassengerId','Survived']].count()
titanic_groupby
```

```python
#여러 개의 aggregation 함수 적용하기, DataFrameGroupby 객체의 agg() 내에 인지로 입력하여 사용
titanic_df.groupby('Pclass')['Age'].agg([max,min])
```

```python
agg_format = {'Age': 'max', 'SibSp': 'sum', 'Fare': 'mean'}
titanic_df.groupby('Pclass').agg(agg_format)
```

```python
#결손 Data 처리하기, NaN 여부 확인 API - isna() , NaN 값을 다른 값으로 대체하는 API - fillna()
titanic_df.isna()
```

```python

titanic_df.isna().sum()
```

```python
#fillna()를 통해 결손 데이터 대체하기
titanic_df['Cabin'] = titanic_df['Cabin'].fillna('C000')  #혹은  titanic_df['cabin'].fillna('C000', inplace = True)
titanic_df.head()
```

```python
#Age 칼럼의 NaN 값을 평균 나이로, Embarked 칼럼의 NaN 값을 'S'로 대체하여 결손데이터 처리하기

titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].mean()) #반환 값을 다시 받는 방법
titanic_df['Embarked'].fillna('S',inplace = True) #inplace 설정을 통해 변경하는 방법
titanic_df.isna().sum()
```

```python
#Apply lambda 식으로 데이터 가공
titanic_df['Name_len'] = titanic_df['Name'].apply(lambda x : len(x))
titanic_df[['Name','Name_len']].head()
```

```python
#lambda 식을 이용, '나이가 15세 미만이면 Child 아니면 Adult'로 구분하는 새로운 칼럼 'Child_Adult' 칼럼을 생성
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x : 'Child' if x < 15 else 'Adult')
titanic_df[['Age','Child_Adult']].head(8)
```

```python
#나이에 따라 세분화된 분류를 수행하는 함수
def get_category(age):
    cat = ''
    if age <= 5: cat = 'Baby'
    elif age <= 12: cat = 'Child'
    elif age <= 18: cat = 'Teenager'
    elif age <= 25: cat = 'Student'
    elif age <= 35: cat = 'Young Adult'
    elif age <= 60: cat = 'Adult'
    else: cat = 'Elderly'
        
    return cat

#lambda 식 위에서 생성한 get_category() 함수를 반환값으로 지정
#get_category(x)는 입력값으로 'Age'칼럼 값으로 받아서 해당하는 cat 반환
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : get_category(x))
titanic_df[['Age', 'Age_cat']].head()
```
